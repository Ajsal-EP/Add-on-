```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array(["A","B","C","D","E"])
y1 = np.array([2,5,9,6,3])
plt.bar(x1,y1,width=0.3,color = 'red')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0,6])
y1 = np.array([0,100])
x2 = np.array([5,8])
y2 = np.array([2,50])

plt.subplot(2,1,1)
plt.plot(x1,y1)
plt.subplot(2,1,2)
plt.plot(x2,y2)

plt.show()
```
